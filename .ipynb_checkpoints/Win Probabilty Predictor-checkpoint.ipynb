## Extracting useful feature 

```python
mf_df=mf_df[['match_id','city','winner','total_runs']]
```

```python
delivery_df=mf_df.merge(df,on='match_id')
```

```python
delivery_df.head(10)
```

- we need to store data about second innings

```python
delivery_df=delivery_df[delivery_df['inning']==2]
```

```python
delivery_df.info()
```

- grouping by on match id and then calculating cumulative sum on each ball

```python
delivery_df['current_score']=delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df
```

- calculating runs left on each ball

```python
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df
```

- Calculating balls left

```python
delivery_df['balls_left']=120-(delivery_df['over']*6+delivery_df['ball'])
```

```python
delivery_df
```

- calculating wickets left after each ball

```python
delivery_df['player_dismissed']
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna(0)
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x:x if x==0 else 1)
delivery_df['player_dismissed']=delivery_df['player_dismissed'].astype('int')
wickets=delivery_df.groupby('match_id')['player_dismissed'].cumsum()

delivery_df['wickets_left']=10-wickets
```

```python
delivery_df
```

```python
delivery_df.groupby('match_id')['player_dismissed'].sum()
```

- calculating current run rate , required run rate , win or lost

```python
delivery_df['crr']=delivery_df['current_score']*6/(120-delivery_df['balls_left'])
```

```python
delivery_df.head()
```

```python
delivery_df['rrr']=(delivery_df['runs_left']*6)/(delivery_df['balls_left'])
```

```python
delivery_df
```

```python
def result(row):
    return 1 if row['batting_team']==row['winner'] else 0
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
numeric_df = delivery_df.select_dtypes(include=['number'])
numeric_df.corr()
```
